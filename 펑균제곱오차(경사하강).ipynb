# 평균제곱오차
## 평균제곱오차와 경사하강을 통한 선형 데이터 예측

```python
import plotly.io as pio
import plotly.graph_objects as go
import numpy as np

pio.templates.default = "plotly_dark"
pio.renderers.default = "notebook_connected"
```

```python
data = [[2,43], [4,55], [6,82], [8,97], [10, 122]]

x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
layout = go.Layout(
    title = 'RAW 데이터',
    xaxis = go.layout.XAxis(title='X축', range=[0, 14]),
    yaxis = go.layout.YAxis(title='Y축', range=[0, 160]),
    height = 600,
    width = 800
)
fig = go.Figure(layout=layout)
fig = fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='raw'))
fig.show()
```

### 평균제곱오차(MSE) 공식:
<img src="img\mse.png" width="50%" height="50%">

### 편미분:
![MSE](img\편미분.png)

```python
x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0

lr = 0.01
epochs = 3000

for i in range(epochs):
    y_pred = a*x_data + b
    error = y_data - y_pred

    a_diff = -(2/len(x_data))*sum((y_data - y_pred)*x_data)
    b_diff = -(2/len(x_data))*sum(y_data - y_pred)

    a = a - lr*a_diff
    b = b - lr*b_diff

    print("epoch={0}, a={1}, b={2}".format(i, a, b))
```

```python
print("a={0}, b={1}".format(a, b))
```
